```python
import csv
f = open('subwaytime.csv', 'r', encoding='utf8')
data = csv.reader(f)
next(data)
next(data)
mx = [0] * 24
mx_station = [''] * 24

for row in data:
    row[4:] = map(int, row[4:])
    for j in range(24):
        a = row[2*j + 4]
        if a > mx[j]:
            mx[j] = a
            mx_station[j] = row[3] + '(' + str(j+4) + '시)'

import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
plt.bar(range(24), mx)
plt.xticks(range(24), mx_station, rotation=90)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('subwaytime.csv', 'r', encoding='utf8')
data = csv.reader(f)
next(data)
next(data)

mx = [0] * 24
mx_station = [''] * 24

for row in data:
    row[4:] = map(int, row[4:])
    for j in range(24):
        b = row[5 + j*2]
        if b > mx[j]:
            mx[j] = b
            mx_station[j] = row[3] + '(' + str(j+4) + ')'

plt.rc('font', family='Malgun Gothic')
plt.bar(range(24), mx, color='b')
plt.xticks(range(24), mx_station, rotation=90)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('subwaytime.csv', 'r', encoding='utf8')
data = csv.reader(f)
next(data)
next(data)


s_in = [0] * 24
s_out = [0] * 24

for row in data:
    row[4:] = map(int, row[4:])
    for i in range(24):
        s_in[i] += row[4 + 2*i]
        s_out[i] += row[5 + 2*i]
    
plt.rc('font', family='Malgun Gothic')
plt.title('지하철 시간대별 승하차 인원 추이')
plt.plot(s_in, label='승차')
plt.plot(s_out, label='하차')
plt.legend()
plt.xticks(range(24), range(4, 28))
plt.show()
```
