## 8월의 최고 기온 데이터 추출

```python
import csv
import matplotlib.pyplot as plt

#%matplotlib qt5
#%matplotlib tk
%matplotlib inline

f_open = open('seoul.csv', 'r', encoding = 'cp949')
data = csv.reader(f_open, delimiter = ',')
header = next(data)
result = []

for row in data :
    if row[-1] != '' :
        if row[0].split('-')[1] == '03' and row[0].split('-')[2] == '28' :
            result.append(float(row[-1]))

plt.plot(result, 'hotpink')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
header = next(data)
high = []  #최고 기온 값을 저장할 리스트 생성
low = []  #최저 기온 값을 저장할 리스트 생성

for row in data :
    if row[-1] != '' and row[-2] != '' :  #최고, 최저 기온 값이 존재한다면
        if 1995 <= int(row[0].split('-')[0]) :  #1995년 이후 데이터라면
            if row[0].split('-')[1] == '03' and row[0].split('-')[2] == '28' :  #3월 28일이라면
                high.append(float(row[-1]))  #최고기온 값 저장
                low.append(float(row[-2]))  #최저 기온 값 저장
                
plt.plot(high, 'hotpink', label = 'HIGH')
plt.plot(low, 'skyblue', label = 'LOW')
plt.legend(loc = 0)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
header = next(data)
avrg = []  #평균 기온 값을 저장할 리스트 생성
years = []

for row in data :
    if row[2] != '' :
        date = row[0].split('-')
        if date[1] == '03' and date[2] == '28' :
            years.append(date[0])
            avrg.append(row[2])
plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.figure(figsize = (20, 5))
plt.title('Average Temperature at 28th March Over the Years')
plt.plot(years, avrg, 'hotpink')
plt.show()
```
