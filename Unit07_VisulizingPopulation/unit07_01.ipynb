```python
import csv
f = open('age.csv')
data = csv.reader(f)
result = []

for row in data :
    if '울산광역시' in row[0] and '성안동' in row[0] :
        for i in row[3:] :
            result.append(i)
print(result)
```

```python
import csv
import matplotlib.pyplot as plt

f = open('age.csv')
data = csv.reader(f)
result = []
for row in data :
    if '신도림' in row[0] :
        for i in row[3:] :
            result.append(i)

plt.figure(figsize = (20, 20))
plt.plot(result)
plt.show()
```
