# 상자 그림 형태 (Boxplot)

```python
import matplotlib.pyplot as plt
import random
result = []
for i in range(13) :
    result.append(random.randint(1, 1000))
print(sorted(result))

plt.boxplot(result)
plt.show()
```

위 그림에서 최댓값, 3/4값, 2/4값(중앙값), 1/4값, 최솟값을 바로 알 수 있다.

다른 위치의 정확한 값을 알고 싶다면 다음 코드를 추가해서 실행해야 한다.  
  
import numpy as np  
result = np.array(result)  
print("1/4: " + str(np.percentile(result, 25)))  
print("2/4: " + str(np.percentile(result, 50)))  
print("3/4: " + str(np.percentile(result, 75)))

```python
import numpy as np
import matplotlib.pyplot as plt
import random

result = []
for i in range(13) :
    result.append(random.randint(1, 1000))
print(sorted(result))

plt.boxplot(result)
plt.show()

result2 = np.array(result)
print("Min: " + str(np.percentile(result2, 0)))
print("1/4: " + str(np.percentile(result2, 25)))
print("2/4: " + str(np.percentile(result2, 50)))
print("3/4: " + str(np.percentile(result2, 75)))
print("Max: " + str(np.percentile(result2, 100)))
```

## 서울 최고 기온 데이터를 boxplot하기

```python
import csv
f = open('seoul.csv')
data = csv.reader(f)
next(data)
result = []

for row in data :
    if row[-1] != '' :
        result.append(float(row[-1]))
    
import matplotlib.pyplot as plt  # import를 사용 직전에 해도 된다.
plt.boxplot(result)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)
aug = []
jan = []

for row in data :
    month = row[0].split('-')[1]
    if row[-1] != '' :
        #if month == '08' :
        #    aug.append(float(row[-1]))
        #if month == '01' :
        #    jan.append(float(row[-1]))
        if month == '01' :
            jan.append(float(row[-1]))
        elif month == '08' :
            aug.append(float(row[-1]))

plt.boxplot(jan)
plt.boxplot(aug)
plt.show()
```

동그라미들은 이상치 (outlier) 값을 표현한 것.

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)
aug = []
jan = []

for row in data :
    month = row[0].split('-')[1]
    if row[-1] != '' :

        if month == '01' :
            jan.append(float(row[-1]))
        elif month == '08' :
            aug.append(float(row[-1]))

plt.boxplot([jan, aug])
plt.show()
```

```python
import matplotlib.pyplot as plt
import csv

f = open('seoul.csv')
data = csv.reader(f)
next(data)

month = [[], [], [], [], [], [], [], [], [], [], [], []]

for row in data :
    if row[-1] != '' :
        month[int(row[0].split('-')[1]) - 1].append(float(row[-1]))
        
plt.figure(figsize = (15, 10))
plt.boxplot(month)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)

day = []
for i in range(31) :
    day.append([])
    
for row in data :
    if row[-1] != '' :
        if row[0].split('-')[1] == '08' :
            day[int(row[0].split('-')[2]) - 1].append(float(row[-1]))

plt.style.use('ggplot')
plt.figure(figsize=(10, 5), dpi=300)
plt.boxplot(day, showfliers=False)

plt.show()
    
```
