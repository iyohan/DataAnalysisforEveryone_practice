#### Using NumPy

```python
import matplotlib.pyplot as plt
import numpy as np
t = np.arange(0., 5., 0.2)
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

#### Without NumPy

```python
import matplotlib.pyplot as plt
t = []
p2 = []
p3 = []
for i in range(0, 50, 2):
    t.append(i / 10)
    p2.append((i / 10) ** 2)
    p3.append((i / 10) ** 3)
plt.plot(t, t, 'r--', t, p2, 'bs', t, p3, 'g^')
plt.show()
```

```python
import matplotlib.pyplot as plt
import random

x = []
y = []
size = []

for i in range(200):
    x.append(random.randint(10, 100))
    y.append(random.randint(10, 100))
    size.append(random.randint(10, 100))
    
plt.scatter(x, y, s=size, c=x, cmap='jet', alpha=0.7)
plt.colorbar()
plt.show()
```

위와 같이 어떤 연산이나 함수를 적용하면 배열의 모든 값이 한꺼번에 계산됨.

```python
import matplotlib.pyplot as plt
import numpy as np
a = np.arange(-np.pi, np.pi, np.pi/100)
plt.plot(a, np.sin(a))
plt.show()
```

```python
a = np.arange(-np.pi, np.pi, np.pi/100)
plt.plot(a, np.sin(a))
plt.plot(a, np.cos(a))
plt.plot(a + np.pi/2, np.sin(a))
plt.show()
```

mask 기능은 어떤 조건에 부합하는 데이터만 선별적으로 저장.

```python
a = np.arange(-5, 5)
print(a)
```

```python
print(a < 0)
```

```python
print(a[a<0])
```

```python
mask1 = abs(a) > 3
print(a[mask1])
```

```python
mask1 = abs(a) > 3
mask2 = abs(a) % 2 == 0
print(a[mask1 + mask2])  # OR
print(a[mask1 * mask2])  # AND
```

## 응용:  numpy 라이브러리를 사용하여 재미있는 버블 차트 그리기

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(-100, 100, 1000)  # 1000개의 랜덤 값 추출
y = np.random.randint(-100, 100, 1000)  # 1000개의 랜덤 값 추출
size = np.random.rand(100) * 100  # 0~1 사이의 실수 100개. 100 스칼라곱 함.
mask1 = abs(x) > 50
mask2 = abs(y) > 50
x = x[mask1 + mask2]  # x[abs(x) > 50 or abs(y) > 50]
y = y[mask1 + mask2]

plt.scatter(x, y, s=size, c=x, cmap='jet', alpha=0.7)
plt.colorbar()
plt.show()
```
