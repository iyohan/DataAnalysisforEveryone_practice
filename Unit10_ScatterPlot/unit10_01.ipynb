# 인구 구조를 산점도 (Scatter Plot)으로 나타내기

## 꺾은선 그래프로 표현하기 (plot()함수)

```python
import csv
f = open('gender.csv')
data = csv.reader(f)
m = []
f = []
name = input('지역 이름을 입력하세요 : ')
print()

for row in data:
    if name in row[0]:
        for i in range(3, 104):
            m.append(int(row[i].replace(',','')))
            f.append(int(row[i+103].replace(',','')))
        break

import matplotlib.pyplot as plt
plt.plot(m, label='Male')
plt.plot(f, label='Female')
plt.legend()
plt.show()
```

## 막대그래프로 표현 (bar() 함수)

```python
import csv
f = open('gender.csv')
data = csv.reader(f)
result = [] 
name = input('지역 이름을 입력하세요 : ')
print()

for row in data:
    if name in row[0]:
        for i in range(3, 104):
            result.append(int(row[i].replace(',','')) - int(row[i+103].replace(',','')))
        break

import matplotlib.pyplot as plt
plt.bar(range(101), result)  #barh()
plt.show()
```

## 연령대별 성별 비율을 산점도로 표현하기

```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv')
data = csv.reader(f)
m = []
f = []
name = input('지역 이름을 입력하세요 : ')
print()
for row in data:
    if name in row[0]:
        for i in range(3, 104):
            m.append(int(row[i].replace(',','')))
            f.append(int(row[i+103].replace(',','')))
        break

plt.scatter(m, f, c=range(101), alpha=0.5, cmap='jet')
plt.colorbar()
plt.plot(range(max(m)), range(max(m)), 'g')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
import math

f = open('gender.csv')
data = csv.reader(f)
m = []
f = []
size = []
name = input('지역 이름을 입력하세요 : ')

for row in data:
    if name in row[0]:
        for i in range(3, 104):
            m_num = int(row[i].replace(',',''))
            f_num = int(row[i+103].replace(',',''))
            m.append(m_num)
            f.append(f_num)
            size.append(math.sqrt(m_num) + math.sqrt(f_num))
        break

plt.style.use('ggplot')
plt.rc('font', family='Malgun Gothic')
plt.figure(figsize=(10, 5), dpi=300)
plt.title(name+' 지역의 성별 인구 그래프')  # + 주위에 띄우는게 나을까?
plt.scatter(m, f, s=size, c=range(101), alpha=0.5, cmap='jet')
plt.colorbar()
plt.plot(range(max(m)), range(max(m)), 'g')
plt.xlabel('남성 인구수')
plt.ylabel('여성 인구수')
plt.show()
```
